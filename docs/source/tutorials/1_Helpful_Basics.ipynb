# Helpful Basics (deprecated)

> **This tutorial has been deprecated.**
>
> *Helpful Basics* was originally written as a Python primer for complete beginners. QSDsan no longer includes a tutorial on Python basics, given the many excellent Python primers available elsewhere.

## Where to go instead

**New to Python or Jupyter?** See *New to Python or Jupyter?* on the [Tutorials index](https://qsdsan.readthedocs.io/en/latest/tutorials/index.html), which links to the official Python tutorial, Python for Everybody, Real Python, Software Carpentry, Scientific Python Lectures, and Project Jupyter.

**Ready to start with QSDsan?** Begin with the [Quick Overview](https://qsdsan.readthedocs.io/en/latest/tutorials/0_Quick_Overview.html), then follow one of the suggested learning paths on the [Tutorials index](https://qsdsan.readthedocs.io/en/latest/tutorials/index.html).

**Looking for the original video walkthrough?** It remains available on our [YouTube channel](https://www.youtube.com/watch?v=g8mXWycdi4E). The content reflects an earlier version of `QSDsan` and is no longer part of the current tutorial series.
